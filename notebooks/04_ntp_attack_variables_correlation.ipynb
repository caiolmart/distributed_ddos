```python
cols = ['flow_key', 'lst_timestamp', 'label'] + \
       ['tt_' + x for x in twotup_cols] + \
       ['ft_' + x for x in fivetup_cols]
```

```python
len(cols)
```

```python
pd.set_option('display.max_columns', 100)
```

```python
data = pd.read_csv('../data/interim/flow_features_DrDoS_NTP.csv', sep=';', names=cols)
print(data.shape)
data.head()
```

```python
y = (data.label == 'DrDoS_NTP') * 1
```

```python
y.shape
```

```python
pearson_corr = dict()
for col in data.columns[3:]:
    X = data[col]
    corr = np.corrcoef(X, y)[0, 1]
    pearson_corr[col] = 0 if np.isnan(corr) else corr
```

```python
sorted_items = sorted(pearson_corr.items(), key=lambda x: abs(x[1]), reverse=True)
names = [x[0] for x in sorted_items]
corrs = [abs(x[1]) for x in sorted_items]
signs = ['Negative' if x[1] < 0 else 'Positive' for x in sorted_items]
```

```python
n_vars = 25
plt.figure(figsize=(8, 5))
plt.title('Pearson Correlation with the Target - NTP Attack - AnubisFlow')
sns.barplot(x=names[:n_vars], y=corrs[:n_vars], hue=signs[:n_vars], dodge=False)
plt.xticks(rotation=90)
plt.ylim([0, 1])
plt.savefig('../plots/correlations/pearson_ourvars_ntp.png', bbox_inches = "tight")
```

## CICFlowMeter Variables

```python
data = pd.read_csv('../data/raw/csv/01-12/DrDoS_NTP.csv')
print(data.shape)
data.head()
```

```python
y = (data.iloc[:, -1] == 'DrDoS_NTP') * 1
```

```python
y.shape
```

```python
pearson_corr = dict()
for col in data.columns[8:-3]:
    #print(col)
    X = data[col]
    corr = np.corrcoef(X, y)[0, 1]
    pearson_corr[col] = 0 if np.isnan(corr) else corr
```

```python
sorted_items = sorted(pearson_corr.items(), key=lambda x: abs(x[1]), reverse=True)
names = [x[0].strip() for x in sorted_items]
corrs = [abs(x[1]) for x in sorted_items]
signs = ['Negative' if x[1] < 0 else 'Positive' for x in sorted_items]
```

```python
n_vars = 25
plt.figure(figsize=(8, 5))
plt.title('Pearson Correlation with the Target - NTP Attack - CICFlowMeter')
sns.barplot(x=names[:n_vars], y=corrs[:n_vars], hue=signs[:n_vars], dodge=False)
plt.ylim([0, 1])
plt.xticks(rotation=90)
plt.savefig('../plots/correlations/pearson_cicvars_ntp.png', bbox_inches = "tight")
```
